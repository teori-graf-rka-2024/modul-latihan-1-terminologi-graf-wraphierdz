Sebuah model bahasa dapat menghasilkan kalimat berdasarkan kata  kata yang ditunjukkan oleh diagram di atas, namun penyusunan kata pada kalimat harus dilakukan secara berurutan berdasarkan tanda panah. Jika ada tanda panah dari kata A ke kata B, artinya di dalam kalimat tersebut, kata B hanya boleh muncul setelah kata A muncul.

Sekarang anda diberikan tantangan untuk menyajikan diagram kata di atas ke dalam daftar ketetanggan di bawah ini :

````python
sentence = "Ubur Ikan Lele Bebek Tegraf Strukdat NLP Is Cool"
sentence = sentence.split()
adj = dict()
# Lengkapi kode rumpang pada fungsi add_connection di bawah ini!
def add_connection(adj:dict, u, v):
    if u not in adj:
        adj[u] = set()
    adj[u].add(v)
    return adj

# Kemudian lanjutkan representasi edge di bawah ini ...
add_connection(adj, sentence[0], sentence[1])
add_connection(adj, sentence[1], sentence[6])
add_connection(adj, sentence[2], sentence[3])
add_connection(adj, sentence[2], sentence[7])
add_connection(adj, sentence[2], sentence[1])
add_connection(adj, sentence[6], sentence[7])
add_connection(adj, sentence[3], sentence[4])
add_connection(adj, sentence[7], sentence[8])
add_connection(adj, sentence[4], sentence[7])
add_connection(adj, sentence[5], sentence[7])
````

Sekarang dilakukan tahap evaluasi / testing pada model bahasa tersebut, anda diberikan input berupa sebuah kata **S** coba anda tentukan kira - kira dengan kata **S**, kalimat terpanjang apa yang bisa dihasilkan?,
** Clue : Degree + Path + Backtracking / Recursive **

````python
S = input()
def longest_sentence(adj, given_word, visited=None):
    if visited is None:
        visited = set()
        
    visited.add(given_word)
    chosen = given_word
    
    for neighbor in adj.get(given_word, {}):
        if neighbor not in visited:
            sentence = longest_sentence(adj, neighbor, visited.copy())
            if len(chosen) < len(sentence):
                chosen = given_word + ' ' + sentence
    return chosen

print(longest_sentence(adj, S))
````

**NOMOR 6 : [Undirected Graph]** <br>
Sebuah kota direpresentasikan sebagai Graf Tidak Berarah dengan persimpangan sebagai node dan jalan sebagai edge. Graf ini disimpan dalam bentuk Adjacency Matrix.

Diberikan kode berikut, lengkapi bagian yang kosong (...) agar program dapat:

* Membaca jumlah N (jumlah simpul/persimpangan).
* Membaca jumlah M (jumlah jalan).
* Membangun Adjacency Matrix untuk graf berdasarkan pasangan simpul yang terhubung.
* Mencetak Adjacency Matrix yang merepresentasikan jaringan jalan.

````python
# Input jumlah simpul (N) dan jumlah jalan (M)
N = int(input())
M = int(input())

# Inisialisasi adjacency matrix dengan 0
graph = [[0 for _ in range(N)] for _ in range(N)]

for _ in range(M):
    u, v = map(int, input().split())
    graph[u][v] = 1  # Karena graf tidak berarah
    graph[v][u] = 1  # Simetri matriks

# Mencetak adjacency matrix
for row in graph:
    print(" ".join(map(str, row)))
````

**NOMOR 7 : [CyberNetic]**

Di sebuah kota futuristik Cybernetic, sistem transportasi telah diotomatisasi menggunakan kecerdasan buatan. Kota ini memiliki N distrik utama, dan beberapa di antaranya dihubungkan oleh jalan berteknologi tinggi yang memungkinkan kendaraan bergerak dengan kecepatan optimal.

Namun, tidak semua jalan memiliki efisiensi yang samasetiap jalan memiliki bobot tertentu, yang merepresentasikan waktu tempuh (dalam detik) dari satu distrik ke distrik lainnya. Jika tidak ada jalan langsung antara dua distrik, maka waktu tempuhnya dianggap tak terhingga (dapat diwakili dengan angka yang sangat besar, misalnya 999999).

Tugas Anda adalah melengkapi kode berikut agar dapat membangun dan mencetak Adjacency Matrix berbobot, yang merepresentasikan jalur di kota ini dengan benar!

````python
# Input jumlah distrik (N) dan jumlah jalan (M)
N = int(input())
M = int(input())

# Inisialisasi adjacency matrix dengan nilai tak terhingga (999999), kecuali diagonalnya 0
graph = [[999999 if i != j else 0 for j in range(N)] for i in range(N)]

# Membaca M jalur jalan dan memperbarui adjacency matrix dengan bobot yang diberikan
for _ in range(M):
    u, v, w = map(int, input().split())  # u dan v adalah distrik yang terhubung, w adalah waktu tempuh
    graph[u][v] = w
    graph[u][v] = w

# Mencetak adjacency matrix berbobot
for row in graph:
    print(" ".join(map(str, row)))
````

**NOMOR 9 : [Temenin Dengklek]** <br>
Di dalam kelas ada N orang bernama 1,2,3,4, ..., N. Dalam rangka keperluan acara kampus, Pak Dengklek mempersiapkan kelompok panitia di mana di dalamnya terdiri dari minimal satu orang dari kelas. Sebuah kelompok Panitia dapat terbentuk jika setiap anggotanya minimal mengenal atau dikenal oleh satu orang lainnya di dalam kelompok atau di kelompok tersebut hanya berisi satu orang saja. Seseorang U bisa saja mengenal orang lainnya yaitu V dengan syarat U  V. Pak Dengklek penasaran jika diberikan informasi beberapa keterangan pasangan (U,V) dari N orang yang ada berapa banyak kelompok panitia minimal yang bisa Pak Dengklek bentuk.

Bantulah ia menghitung banyak kelompok minimal yang dapat Pak Dengklek bentuk!
<br><br>
**Format Masukan** <br>
Baris pertama berisi bilangan bulat N, dan Q.
Q baris berikutnya masing  masing berisikan Ui dan Vi yang menyatakan bahwa Ui mengenal Vi untuk (1 < i < Q)
<br><br>
**Format Keluaran** <br>
Keluarkan satu baris jawaban berupa banyaknya kelompok minimal yang dapat dibentuk.
<br><br>
**Contoh Masukan dan Keluaran** <br>
Contoh Masukan
```
5 3
1 2
2 3
4 5
```
Contoh Keluaran
```
2
```


Penjelasan Contoh :
Pak Dengklek bisa membentuk 2 kelompok yaitu kelompok 1 : [1,2,3] dan kelompok 2 : [4,5].


````python
n, q = map(int, input().split())

graph = {}
for i in range(q):
    u,v = map(int, input().split())

    if u not in graph:
        graph[u] = set()
    graph[u].add(v)

    if v not in graph:
        graph[v] = set()
    graph[v].add(u)

    visited = set()
    jmlGroup = 0

    # DFS
    for i in range(1, n + 1):
        if i not in visited:
            stack = [i]
            visited.add(i)
            while stack:
                node = stack.pop()
                if node in graph:
                    for neighbor in graph[node]:
                        if neighbor not in visited:
                            visited.add(neighbor)
                            stack.append(neighbor)
            jmlGroup += 1
            
print(jmlGroup)
````
